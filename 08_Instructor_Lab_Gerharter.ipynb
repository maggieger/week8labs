## Sorting and Ordering data

#### Select columns Team_type, human_work, work_per_human

```python
gh_teams.filter(["Team_type", "human_work", "work_per_human"])
```

#### Select columns that end in t using regex

```python
(gh_teams.filter(regex="t$"))
```

#### Sort descending by Team_size_class, human_work, work_per_human

```python
(gh_teams.sort_values(by=["Team_size_class", "human_work", "work_per_human"], ascending=False))
```

#### Select human-bot teams with bot_members_count >= 3

```python
(gh_teams
 .query("Team_type == 'human-bot' & bot_members_count >= 3"))
```

####  Select human teams that are Large and human_gini >= 0.75

```python
(gh_teams
 .query("Team_type == 'human' & Team_size_class == 'Large' & human_gini >= 0.75"))
```

#### How many teams are Small or Large?

```python
(gh_teams
 .query("Team_size_class in ['Small', 'Large']")
 .shape[0])
```

#### How many Small or Large teams have human_gini <= 0.25?

```python
(gh_teams
 .query("Team_size_class in ['Small', 'Large'] & human_gini <= 0.25")
 .shape[0])
```

#### How many human teams are in Medium category?

```python
(gh_teams
 .query("Team_type =='human' & Team_size_class == 'Medium'")
 .shape[0])
```
