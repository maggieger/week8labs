## Data Wrangling

### Mean of `user_commits`

### Columns that start with 'p' or contain 'g' in `turnover_new`

```python
turnover_new = turnover.filter(regex = "(^p)|(g)")
turnover_new
```

### `project_commits`> 9000 in `turnover_new`

```python
turnover_new.query("project_commits > 9000")
```

### Combination of the last 2 steps

```python
turnover_newer = (turnover
                  .filter(regex = "(^p)|(g)")
                  .query("project_commits > 9000"))
turnover_newer
```

### ends with `e` or contain `s`, and rows where `gh_tenure >= 1991`

```python
(turnover
 .filter(regex="(e$)|(s)")
 .query("gh_tenure >= 1991"))
```

### Mean and Median of `user_commits`

```python
(turnover
 .filter(regex="(e$)|(s)|(^user_commits$)")
 .query("gh_tenure >= 1991")
 .user_commits
 .agg(["mean","median"]))
```

#### Analysing `user_commits` stats  

When analysing all users:  
- Mean: 339.36575052854124  
- Median: 18.0

When only analysing users with 'gh_tenure >= 1991' and from columns that end with either an 'e' or contain an 's':
- Mean: 3.5
- Median: 3.5

This comparison suggests that the filtered group (users who have been on GitHub longer) are less active when it comes to commits than the overall group. The overall dataset includes users with a large number of commits, which raises the mean.
